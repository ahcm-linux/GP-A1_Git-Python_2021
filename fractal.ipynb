## Projeto em Python

Código que cria triângulos de Siérpinski, um exemplo clássico da geometria fractal.

### Algoritmo:

Execute os seguintes passos em sequência:
1. Defina um conjunto com três pontos que determinam os vértices de um triângulo isóceles
2. Escolha aleatóreamente um ponto interno ao triângulo
3. Escolha aleatoreamente um dos vértices do triângulo
4. Obtenha o ponto médio entre os pontos selecionados nos passos 2 e 3
5. Repita N vezes os passos 3 e 4, em sequência

Para N grande, será gerada uma figura fractal conhecida como triângulo de Siérpinski.

```python
import matplotlib.pyplot as plt
import random as rd
import math
```

```python
x = [0, 2, 1]
y = [0, 0, math.sqrt(3)]

rx = 1
ry = 1.5

plt.scatter(x, y, s=20, c="red")
plt.scatter(rx, ry, s=20, c="green")

N = 1000
for i in range(1, N + 1):
    selection = rd.randint(0, len(x) - 1)
    rx = (rx + x[selection]) / 2
    ry = (ry + y[selection]) / 2
    plt.scatter(rx, ry, s = 1, c="black")

plt.show()
```
